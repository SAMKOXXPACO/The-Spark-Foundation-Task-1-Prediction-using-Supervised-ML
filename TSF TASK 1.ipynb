```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics  
df=pd.read_csv("C:/Users/hp/Desktop/Data Science/TSF/task1 supervised ML/student_scores - student_scores.csv")
df
```

```python
df.describe()
```

```python
# Plotting the distribution of scores
df.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

```python
from sklearn.linear_model import Ridge,Lasso,LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
x=pd.DataFrame(df["Hours"])
y=pd.DataFrame(df["Scores"])
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.40,random_state=0)
```

# LINEAR REGRESSION

```python
lm_LR=LinearRegression()
lm_LR.fit(x_train,y_train)
```

```python
Pred_LR=lm_LR.predict(x_test)
```

```python
r2_score(Pred_LR,y_test)
```

```python
print("Linear Regression student score predictions who studies for 9.25 hours is:",lm_LR.predict([[9.25]]))
```

```python
print('Mean Absolute Error using Linear Regression:', 
      metrics.mean_absolute_error(y_test, Pred_LR)) 
```

# LASSO REGRESSION

```python
lm_Lasso=Lasso()
lm_Lasso.fit(x_train,y_train)
```

```python
Pred_Lasso=lm_Lasso.predict(x_test)
```

```python
r2_score(Pred_Lasso,y_test)
```

```python
print("Lasso Regression student score predictions who studies for 9.25 hours is:",lm_Lasso.predict([[9.25]]))
```

```python
print('Mean Absolute Error using Lasso Regression:', 
      metrics.mean_absolute_error(y_test, Pred_Lasso)) 
```

# RIDGE REGRESSION

```python
lm_Ridge=Ridge()
lm_Ridge.fit(x_train,y_train)
```

```python
Pred_Ridge=lm_Ridge.predict(x_test)
```

```python
r2_score(Pred_Ridge,y_test)
```

```python
print("Ridge regression student score predictions who studies for 9.25 hours is:",lm_Ridge.predict([[9.25]]))
```

```python
print('Mean Absolute Error using Ridge Regression:', 
      metrics.mean_absolute_error(y_test, Pred_Ridge)) 
```
